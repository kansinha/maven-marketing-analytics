```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('marketing_data.csv')
data
```

```python
data.columns
data.info()
data.head()
data.tail()
```

```python
data.columns = data.columns.str.strip()
print(data.isnull().sum())
```

# income column has missing values so we will impute with median income

```python
mean_income= data['Income'].mean()
mean_income
data['Income'].fillna(mean_income, inplace=True)
print(data.isnull().sum())
```

# Calculated KPI's

```python
# Totalspend
data['Total_spend'] = (data['MntWines'] + data['MntFruits'] 
                       + data['MntMeatProducts'] + data['MntFishProducts'] + 
                       data['MntSweetProducts'] + data['MntGoldProds'])
data['Total_spend']

# Totalpurchase
data['Total_purchase'] = (data['NumWebPurchases']+ data['NumCatalogPurchases'] + 
                          data['NumStorePurchases'])

# SpendperPurchase
data['Spend/Purchase'] = (data['Total_spend'] / data['Total_purchase'].replace(0, 1)).round(2)

# Campaigns accepted
data['CampaignsAccepted'] = (data['AcceptedCmp1'] + data['AcceptedCmp2'] + 
                              data['AcceptedCmp3'] + data['AcceptedCmp4'] + 
                              data['AcceptedCmp5'] + data['Response'])

#verify new columns
print(data[['Total_spend','Total_purchase',
            'Spend/Purchase','CampaignsAccepted']].describe().round(2))
```

```python
# Ghost customers. Customers who spent 0
print(data[data['Total_purchase'] == 0].shape[0], "customers with 0 purchases")


# Top 5 spenders
print(data.nlargest(5, 'Total_spend')[['Income','Total_spend','CampaignsAccepted']])
```

## Insights
